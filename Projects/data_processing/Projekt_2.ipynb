```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# url för steam sidans special offers
url = 'https://store.steampowered.com/search/?specials=1&page=1'

# Skicka request och få content med hjälp av buutifulsuup
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# En tom lista för data
data = []

# Hitta alla game listing element
games = soup.find_all('a', class_='search_result_row')

# Loopa igenom element/klassar för att hitta rätta saker
for game in games:
    try:
        # För att få spelnamn och om inte hittar så lägger värdet n/a
        name_element = game.find('span', class_='title')
        if name_element:
            #strip gör av me whitespace
            name = name_element.text.strip()
        else:
            name = 'N/A'   
     
        # för att få rating
        rating_element = game.find('span', class_='search_review_summary')
        #Fånga det rätta värdet med att splitta br och välja första indexet
        rating = rating_element['data-tooltip-html'].split('<br>')[0] 
        if rating_element:
            #strip gör av me whitespace
            rating = name_element.text.strip()
        else:
            rating = 'N/A' 
        
        # För att få hur många reviews och om inte hittar så returnerar det 0
        reviews = rating_element['data-tooltip-html'].split('<br>')[1].split(' ')[0] if rating_element else '0'
        
        # För att få rabatten och om inte hittar ett discount_element så returnerar det 0
        discount_element = game.find('div', class_='discount_pct')
        discount = discount_element.text.strip().replace('%', '') if discount_element else '0'
        
        # för att få priset efter rabatten och om inte hittar elementet så returnerar det n/a
        final_price_element = game.find('div', class_='discount_final_price')
        final_price = final_price_element.text.strip() if final_price_element else 'N/A'
        
        # För att få priset före rabatten och om inte hittar elementet så returnerar det n/a
        original_price_element = game.find('div', class_='discount_original_price')
        original_price = original_price_element.text.strip() if original_price_element else 'N/A'
        
        # För att få release datum för spelet om om inte hittar elementet så returerar det n/a
        release_year_element = game.find('div', class_='col search_released responsive_secondrow')
        release_year = release_year_element.text.strip() if release_year_element else 'N/A'

        # Logik för att få operativsystemet. Om hittar klassen t.ex. windows(win) så värdet 1(true) annars 0(false)
        win = 1 if game.find('span', class_='platform_img win') else 0
        lin = 1 if game.find('span', class_='platform_img linux') else 0
        osx = 1 if game.find('span', class_='platform_img mac') else 0
        
        # Appenda datan/varaiablerna till den tomma data listan
        data.append([name, rating, reviews, discount, final_price, original_price, release_year, win, lin, osx])
        
    except Exception as e:
        print(f"Err: {e}")
        continue

# Gör till dataframe och lägg till kolumner/namn 
df = pd.DataFrame(data, columns=['Spelnamn', 'Rating', '#Reviews', 'Rabatt%', 'SlutPris', 'OrdinariePris', 'Utgivningsår', 'Win', 'Lin', 'OSX'])

# Spara till csv
df.to_csv('steam_specials.csv', index=False, encoding='utf-8')



```
